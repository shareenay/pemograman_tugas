Studi Kasus: Digitalisasi Seleksi Beasiswa Universitas Trululala

Latar Belakang: Universitas Trululala setiap tahunnya mengalokasikan dana bantuan pendidikan untuk ratusan mahasiswa. Selama ini, proses seleksi dilakukan secara manual menggunakan formulir kertas. Hal ini menyebabkan beberapa masalah:

    1. Antrean Panjang: Petugas kewalahan menghitung satu per satu kriteria mahasiswa.
    2. Ketidaktelitian: Sering terjadi kesalahan input di mana mahasiswa dengan IPK rendah tetap masuk ke tahap seleksi berkas.
    3. Rekapitulasi Lambat: Bagian keuangan sulit mengetahui jumlah total penerima beasiswa secara cepat di akhir hari.

Tujuan Program: Program harus bisa berjalan terus-menerus hingga petugas mengetik kata kunci 'selesai'. Di akhir sesi, program wajib menampilkan laporan ringkas tentang berapa banyak mahasiswa yang mendaftar, berapa yang berhasil lolos di masing-masing kategori, dan berapa banyak yang gagal dalam seleksi tersebut.

```python
total_pendaftar = 0
lolos_prestasi = 0
lolos_bantuan = 0

print("*************************************")
print("  SISTEM SELEKSI BEASISWA TRULULALA")
print("*************************************")

while True:
    print("\nData mahasiswa ke-", total_pendaftar+1)
    nama = input("Nama Mahasiswa (ketik 'selesai' untuk ringkasan)")

    if nama =='selesai':
        break
    
    ipk = float(input("Masukkan IPK (skala 4.0): "))
    
    if ipk < 3.0:
        print("Status : Maaf, IPK tidak memenuhi syarat")

    else:
        print("Pilih Kategori Beasiswa: ")
        print("1. Beasiswa Prestasi")
        print("2. Beasiswa Bantuan")
        pilihan = input("masukan pilihan 1 atau 2")

        if pilihan == "1":
            sertifikat = int(input("Berapa Jumlah Sertifikat Lomba Nasional: "))
            if ipk >= 3.5 and sertifikat >= 1:
                print("status :", nama, "Dinyatakan" + " " + "\033[92mLULUS\033[0m" + " " + "Beasiswa Prestasi")
                lolos_prestasi += 1

            else:
                print("status :", nama, "Dinyatakan" + " " + "\033[91mTIDAK LULUS\033[0m" + " " + "Mendapatkan Beasiswa Prestasi")
        
        elif pilihan =="2":
            gaji = int(input("Masukan Penghasilan Orang tua: Rp."))
            if gaji < 4000000:
                print("Status :", nama, "Dinyatakan" + " " + "\033[92mLULUS\033[0m" + " " + "Beasiswa Bantuan")
                lolos_bantuan += 1
            
            else:
                print("Status :", nama, "Dinyatakan" + " " + "\033[91mTIDAK LULUS\033[0m" + " " + "Mendapatkan Beasiswa Bantuan")
        else:
            print("pilihan tidak valid")
            continue

        total_pendaftar += 1

print("****************************")
print("       RINGKASAN        ")
print("****************************")
print("Total Mahasiswa yang diproses :", total_pendaftar)
print("Lolos Beasiswa Prestasi       :", lolos_prestasi)
print("Lolos Beasiswa Bantuan        :", lolos_bantuan)
print("Total Tidak Lolos             :", total_pendaftar - (lolos_prestasi + lolos_bantuan))                            
```
